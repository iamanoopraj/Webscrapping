```python
import requests
from bs4 import BeautifulSoup
```

```python
url = r'https://quotes.toscrape.com/'

page = requests.get(url)

print(page)
```

```python
soup = BeautifulSoup(page.text,'lxml')

print(soup.prettify())
```

```python
box = soup.find('div',class_='quote')
```

```python
print(box)
```

```python
quote = box.find('span',class_='text').text
print(quote)
```

```python
author = box.find('small',class_='author').text
print(author)
```

```python
scrapped_quotes = {'quotes':[],
                  'authors':[],
                  'links':[]}
```

```python
boxes = soup.find_all('div',class_='quote')

for box in boxes:
    quote = box.find('span',class_='text').text
    author = box.find('small',class_='author').text
    author_link = url+box.find('a')['href']
    
    scrapped_quotes['quotes'].append(quote)
    scrapped_quotes['authors'].append(author)
    scrapped_quotes['links'].append(author_link)
    
```

```python
scrapped_quotes
```

```python
import pandas as pd

pd.DataFrame(data=scrapped_quotes)
```
